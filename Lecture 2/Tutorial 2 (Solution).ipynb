# Tutorial 2: Plotting, Computing and Data Analysis

## 1. Plotting

Plotting is a very important tool in data analysis. It helps us to visualize the data and understand the underlying patterns. In this tutorial, we will use the `matplotlib` library to plot some data.

### 1.1 Import matplotlib

```python
# !pip install matplotlib if not installed
import matplotlib.pyplot as plt
```

### 1.2 Line Plot

Let's plot a simple line plot using the `plot` function. We will plot the function $y = x^2$ for $x$ in the range $[-10, 10]$.

```python
x = list(range(-10, 11))
y = [i**2 for i in x]
plot = plt.plot(x, y)
plt.show()
```

In case we want to plot multiple lines on the same plot, we can call the `plot` function multiple times before calling `show`.

```python
y2 = [i**3 for i in x]
plot1 = plt.plot(x, y)
plot2 = plt.plot(x, y2)
plt.show()
```

We can also customize the plot by adding labels, title, and legend.

```python
plot1 = plt.plot(x, y, label='y = x^2')
plot2 = plt.plot(x, y2, label='y = x^3')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Functions')
plt.legend()
plt.show()
```

### 1.3 Scatter Plot

Scatter plots are used to visualize the relationship between two variables. Let's plot a scatter plot of $y = x^2$ for $x$ in the range $[-10, 10]$.

```python
x = list(range(-10, 11))
y = [i**2 for i in x]
plot = plt.scatter(x, y)
plt.show()
```

### 1.4 Bar Plot

Bar plots are used to compare different categories of data. Let's plot a bar plot of the number of students in different classes.

```python
classes = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5']
students = [20, 25, 30, 15, 10]
plot = plt.bar(classes, students)
plt.show()
```

### 1.5 Histogram

Histograms are used to visualize the distribution of a continuous variable. Let's plot a histogram of the ages of students.

```python
import random # it may be better to import it at the beginning of the notebook
ages = [random.randint(18, 25) for i in range(50)]
plot = plt.hist(ages, bins=range(18, 26))
plt.show()
```

### 1.6 Pie Chart

Pie charts are used to visualize the relative sizes of different categories. Let's plot a pie chart of the number of students in different classes.

```python
classes = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5']
students = [20, 25, 30, 15, 10]
plot = plt.pie(students, labels=classes, autopct='%1.1f%%')
plt.show()
```

## 2. Computing

### 2.1 Basic Operations

With Computing we define the calculation of derivatives, integrals, limits, differential equations and others useful tools to solve problems in mathematics and physics.

### 2.2 Derivatives

Derivatives are used to calculate the rate of change of a function. Let's calculate the derivative of the function $y = x^2$.

```python
import sympy as sp
x = sp.symbols('x')
y = x**2
dy_dx = sp.diff(y, x)
dy_dx
```

### 2.3 Integrals

Integrals are used to calculate the area under the curve of a function. Let's calculate the integral of the function $y = x^2$.

```python
y = x**2
integral = sp.integrate(y, x)
integral
```

### 2.4 Limits

Limits are used to calculate the value that a function approaches as the input approaches a certain value. Let's calculate the limit of the function $y = x^2$ as $x$ approaches 0.

```python
y = x**2
limit = sp.limit(y, x, 0)
limit
```

If we want to calculate the limit of a function as $x$ approaches infinity, we can use `sp.oo` to represent infinity.

```python
y = x**2
limit = sp.limit(y, x, sp.oo)
limit
```

### 2.5 Differential Equations

Differential equations are used to model the rate of change of a function. Let's solve the differential equation $dy/dx = 2x$.

```python
y = sp.Function('y')
dy_dx = 2*x
solution = sp.dsolve(sp.Derivative(y(x), x) - dy_dx)
solution
```

## 3. Data Analysis

### 3.1 Importing Data

Data analysis is the process of cleaning, transforming, and modeling data to discover useful information. In this tutorial, we will use the `pandas` library to analyze some data.

### 3.2 Loading Data

Let's load a dataset containing information about students.

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [20, 21, 22, 23, 24]}
```

```python
df = pd.DataFrame(data)
df
```

DataFrames are two-dimensional data structures with labeled axes (rows and columns). We can access the rows and columns of a DataFrame using the `loc` and `iloc` methods.

```python
row = df.loc[0]
column = df['Name']
print(row)
print(column)
```

### 3.3 Descriptive Statistics

Descriptive statistics are used to summarize the main characteristics of a dataset. Let's calculate the mean, median, and standard deviation of the ages of students.

```python
mean = df['Age'].mean()
median = df['Age'].median()
std = df['Age'].std()
print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Standard Deviation: {std}')
```
